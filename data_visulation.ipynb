```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips= sns.load_dataset('tips')
```

```python

tips.info()
tips.isnull().sum()
```

```python
cat_list = ["sex", "smoker", "time", "day"]
num_list = ["total_bill", "tip", "size"]
```

```python
# Günlere göre toplam hesap tutarını hesapla
daily_total_bill = tips.groupby('day')['total_bill'].sum()

# Çubuk grafik oluştur
plt.bar(daily_total_bill.index, daily_total_bill.values, color='skyblue')
plt.xlabel('Gün')
plt.ylabel('Toplam Hesap Tutarı')
plt.title('Günlere Göre Toplam Hesap Tutarı')
plt.show()
```

```python

average_tip_by_sex = tips.groupby('sex')['tip'].mean().reset_index()

# Bar grafik oluştur
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='tip', data=average_tip_by_sex, palette='husl')
plt.xlabel('Cinsiyet')
plt.ylabel('Ortalama Bahşiş Miktarı')
plt.title('Cinsiyet Kırılımında Ortalama Bahşiş Miktarı')
plt.show()
```

```python

# Histogram oluştur
plt.hist(tips['total_bill'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Toplam Hesap Tutarı')
plt.ylabel('Frekans')
plt.title('Toplam Hesap Tutarı Histogramı')
plt.show()
```

```python
# Countplot oluştur
sns.countplot(x='sex', data=tips, palette='husl')
plt.xlabel('Cinsiyet')
plt.ylabel('Frekans')
plt.title('Cinsiyet Frekansları')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 'tips' veri setini yükle
tips = sns.load_dataset('tips')

# Cinsiyete göre sigara içme durumu gösteren countplot oluştur
sns.countplot(x='sex', hue='smoker', data=tips, palette='Set2')
plt.xlabel('Cinsiyet')
plt.ylabel('Frekans')
plt.title('Cinsiyete Göre Sigara İçme Durumu')
plt.legend(title='Sigara İçme Durumu', labels=['Hayır', 'Evet'])
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], color='skyblue', edgecolor='black')
plt.xlabel('Toplam Hesap Tutarı')
plt.ylabel('Bahşiş Miktarı')
plt.title('Toplam Hesap Tutarı ve Bahşiş Miktarı İlişkisi')
plt.show()
```

```python
# Cinsiyete göre hem ortalama hesap tutarı hem de ortalama bahşiş miktarını gösteren barplot oluştur
sns.barplot(x='sex', y='total_bill', data=tips, ci=None, color='skyblue', label='Ortalama Hesap Tutarı')
sns.barplot(x='sex', y='tip', data=tips, ci=None, color='orange', label='Ortalama Bahşiş Miktarı')
plt.xlabel('Cinsiyet')
plt.ylabel('Ortalama Miktar')
plt.title('Cinsiyete Göre Ortalama Hesap Tutarı ve Bahşiş Miktarı')
plt.legend()
plt.show()
```

```python
# Korelasyon matrisini oluştur
correlation_matrix = tips.corr()

# Korelasyon matrisini görselleştir
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Değişkenler Arasındaki Korelasyon')
plt.show()
```

```python
# Bahşiş miktarını içeren boxplot oluştur
plt.figure(figsize=(8, 6))
sns.boxplot(x='tip', data=tips, color='skyblue')
plt.xlabel('Bahşiş Miktarı')
plt.title('Bahşiş Miktarı Boxplot')
plt.show()
```

```python
contingency_table = pd.crosstab(tips['day'], tips['sex'])
sns.heatmap(contingency_table, annot=True, cmap='YlGnBu')
plt.title('Cinsiyet ve Gün İlişkisi (Heatmap)')
plt.show()
```
